# 🌍 World Happiness Report Analysis (2015–2019)
### Charts included: Bar Chart | Line Chart | Scatter Plot | Pie Chart | Horizontal Bar Chart | Heatmap

## Step 1 — Import Libraries 📦

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print('✅ Done!')
```

## Step 2 — Load All 5 Files 📂

```python
df2015 = pd.read_csv('2015.csv')
df2016 = pd.read_csv('2016.csv')
df2017 = pd.read_csv('2017.csv')
df2018 = pd.read_csv('2018.csv')
df2019 = pd.read_csv('2019.csv')

print('Files loaded!')
```

## Step 3 — Clean & Rename Columns 🧹

Every year has different column names. We rename them to match.

```python
# Rename 2015 columns
df2015 = df2015.rename(columns={
    'Country'                       : 'Country',
    'Happiness Rank'                : 'Rank',
    'Happiness Score'               : 'Score',
    'Economy (GDP per Capita)'      : 'GDP',
    'Family'                        : 'Social_Support',
    'Health (Life Expectancy)'      : 'Life_Expectancy',
    'Freedom'                       : 'Freedom',
    'Trust (Government Corruption)' : 'Corruption',
    'Generosity'                    : 'Generosity'
})
df2015['Year'] = 2015

# Rename 2016 columns
df2016 = df2016.rename(columns={
    'Country'                       : 'Country',
    'Happiness Rank'                : 'Rank',
    'Happiness Score'               : 'Score',
    'Economy (GDP per Capita)'      : 'GDP',
    'Family'                        : 'Social_Support',
    'Health (Life Expectancy)'      : 'Life_Expectancy',
    'Freedom'                       : 'Freedom',
    'Trust (Government Corruption)' : 'Corruption',
    'Generosity'                    : 'Generosity'
})
df2016['Year'] = 2016

# Rename 2017 columns
df2017 = df2017.rename(columns={
    'Country'                          : 'Country',
    'Happiness.Rank'                   : 'Rank',
    'Happiness.Score'                  : 'Score',
    'Economy..GDP.per.Capita.'         : 'GDP',
    'Family'                           : 'Social_Support',
    'Health..Life.Expectancy.'         : 'Life_Expectancy',
    'Freedom'                          : 'Freedom',
    'Trust..Government.Corruption.'    : 'Corruption',
    'Generosity'                       : 'Generosity'
})
df2017['Year'] = 2017

# Rename 2018 columns
df2018 = df2018.rename(columns={
    'Country or region'             : 'Country',
    'Overall rank'                  : 'Rank',
    'Score'                         : 'Score',
    'GDP per capita'                : 'GDP',
    'Social support'                : 'Social_Support',
    'Healthy life expectancy'       : 'Life_Expectancy',
    'Freedom to make life choices'  : 'Freedom',
    'Perceptions of corruption'     : 'Corruption',
    'Generosity'                    : 'Generosity'
})
df2018['Year'] = 2018

# Rename 2019 columns
df2019 = df2019.rename(columns={
    'Country or region'             : 'Country',
    'Overall rank'                  : 'Rank',
    'Score'                         : 'Score',
    'GDP per capita'                : 'GDP',
    'Social support'                : 'Social_Support',
    'Healthy life expectancy'       : 'Life_Expectancy',
    'Freedom to make life choices'  : 'Freedom',
    'Perceptions of corruption'     : 'Corruption',
    'Generosity'                    : 'Generosity'
})
df2019['Year'] = 2019

# Keep only the columns we need
my_columns = ['Country', 'Rank', 'Score', 'GDP', 'Social_Support',
              'Life_Expectancy', 'Freedom', 'Corruption', 'Generosity', 'Year']

df2015 = df2015[my_columns]
df2016 = df2016[my_columns]
df2017 = df2017[my_columns]
df2018 = df2018[my_columns]
df2019 = df2019[my_columns]

# Stack all 5 years into ONE big table
df = pd.concat([df2015, df2016, df2017, df2018, df2019], ignore_index=True)

print('Total rows:', len(df))
df.head()
```

## Step 4 — Quick Data Check 🔍

```python
# Basic statistics
df.describe().round(2)
```

```python
# Check missing values
print('Missing values:')
print(df.isnull().sum())
```

# 📊 VISUALIZATIONS
---

## Chart 1 — BAR CHART 📊
### Top 10 Happiest Countries in 2019

```python
# Get 2019 data only
data_2019 = df[df['Year'] == 2019]

# Sort and take top 10
top10 = data_2019.sort_values('Score', ascending=False).head(10)

# Draw bar chart
plt.figure(figsize=(12, 6))
plt.bar(top10['Country'], top10['Score'], color='steelblue', edgecolor='white')

# Labels and title
plt.title('Top 10 Happiest Countries in 2019', fontsize=15, fontweight='bold')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45, ha='right')  # rotate country names so they fit

plt.tight_layout()
plt.savefig('chart1_bar.png')
plt.show()
print('✅ Bar Chart saved!')
```

## Chart 2 — LINE CHART 📈
### How Did Average Happiness Change Each Year?

```python
# Calculate the average happiness score for each year
avg_per_year = df.groupby('Year')['Score'].mean()

# Draw line chart
plt.figure(figsize=(10, 5))
plt.plot(avg_per_year.index, avg_per_year.values,
         marker='o',        # add circle dots on each year
         color='green',
         linewidth=2.5,
         markersize=9)

# Add score labels on each dot
for year, score in zip(avg_per_year.index, avg_per_year.values):
    plt.text(year, score + 0.03, f'{score:.2f}', ha='center', fontsize=11)

plt.title('Average World Happiness Score Each Year', fontsize=15, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.xticks([2015, 2016, 2017, 2018, 2019])

plt.tight_layout()
plt.savefig('chart2_line.png')
plt.show()
print('✅ Line Chart saved!')
```

## Chart 3 — SCATTER PLOT 🔵
### Does More Money = More Happiness? (GDP vs Happiness Score)

```python
# Draw scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'],          # x-axis = GDP
            df['Score'],        # y-axis = Happiness Score
            color='orange',
            alpha=0.5,          # make dots a bit transparent
            s=40)               # size of dots

plt.title('GDP per Capita vs Happiness Score', fontsize=15, fontweight='bold')
plt.xlabel('GDP per Capita (Wealth of the country)')
plt.ylabel('Happiness Score')

plt.tight_layout()
plt.savefig('chart3_scatter.png')
plt.show()
print('✅ Scatter Plot saved!')
```

## Chart 4 — PIE CHART 🥧
### What Percentage of Countries Fall in Each Happiness Group? (2019)

```python
# Split countries into 3 groups based on their happiness score
def happiness_group(score):
    if score >= 6.0:
        return 'Happy (Score ≥ 6)'
    elif score >= 4.0:
        return 'Moderate (4 to 6)'
    else:
        return 'Unhappy (Score < 4)'

# Apply the function to 2019 data
data_2019 = df[df['Year'] == 2019].copy()
data_2019['Group'] = data_2019['Score'].apply(happiness_group)

# Count how many countries in each group
group_counts = data_2019['Group'].value_counts()

# Draw pie chart
plt.figure(figsize=(8, 8))
plt.pie(group_counts.values,
        labels=group_counts.index,
        autopct='%1.1f%%',       # show percentage on each slice
        colors=['#2ecc71', '#f39c12', '#e74c3c'],
        startangle=140,
        explode=[0.05, 0.05, 0.05])  # slightly separate the slices

plt.title('Countries by Happiness Group in 2019', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.savefig('chart4_pie.png')
plt.show()
print('✅ Pie Chart saved!')
```

## Chart 5 — HORIZONTAL BAR CHART (hbar) 📊
### Which Factor Contributes Most to Happiness?

```python
# Calculate the average value of each factor in 2019
factors = ['GDP', 'Social_Support', 'Life_Expectancy', 'Freedom', 'Generosity', 'Corruption']

avg_factors = df[df['Year'] == 2019][factors].mean()

# Sort from smallest to largest
avg_factors = avg_factors.sort_values()

# Draw horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(avg_factors.index,    # factor names on y-axis
         avg_factors.values,   # values on x-axis
         color='mediumpurple',
         edgecolor='white')

# Add value labels at the end of each bar
for i, value in enumerate(avg_factors.values):
    plt.text(value + 0.005, i, f'{value:.3f}', va='center', fontsize=11)

plt.title('Average Factor Contribution to Happiness (2019)', fontsize=15, fontweight='bold')
plt.xlabel('Average Value')

plt.tight_layout()
plt.savefig('chart5_hbar.png')
plt.show()
print('✅ Horizontal Bar Chart saved!')
```

## Chart 6 — HEATMAP 🌡️
### How Are All Factors Related to Each Other?

```python
# Choose columns to compare
all_cols = ['Score', 'GDP', 'Social_Support', 'Life_Expectancy', 'Freedom', 'Corruption', 'Generosity']

# Calculate correlation (values close to 1 = strongly related)
corr_table = df[all_cols].corr().round(2)

# Draw heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(corr_table,
            annot=True,         # show numbers inside boxes
            fmt='.2f',          # 2 decimal places
            cmap='coolwarm',    # red = positive, blue = negative
            linewidths=0.5,
            vmin=-1, vmax=1)    # scale from -1 to 1

plt.title('Correlation Heatmap — How Factors Relate to Each Other', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('chart6_heatmap.png')
plt.show()
print('✅ Heatmap saved!')
```

---
## Step 5 — Correlation Analysis 📈
### Which factor has the strongest link with happiness?

```python
# Correlation of each factor with Happiness Score
corr_with_happiness = df[all_cols].corr()['Score'].drop('Score')
corr_with_happiness = corr_with_happiness.sort_values(ascending=False)

print('Factor vs Happiness Score Correlation:')
print('-' * 40)
for factor, value in corr_with_happiness.items():
    print(f'{factor:<20} : {value:.3f}')

print()
print('👑 Most related factor :', corr_with_happiness.idxmax())
print('📉 Least related factor:', corr_with_happiness.idxmin())
```
